```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv("Corona_NLP_train.csv", encoding='latin-1')
df_test = pd.read_csv("Corona_NLP_test.csv", encoding='latin-1')
```

```python
df = pd.concat([df_train, df_test], axis=0, ignore_index=True)
```

```python
df
```

```python
df['len'] = df['OriginalTweet'].apply(lambda x: len(x.split()))
```

```python
df
```

```python
df[df['Sentiment'] == "Extremely Positive"]
```

```python
# plot density of tweet length
sns.kdeplot(df[df['Sentiment'] == "Extremely Positive"]['len'], shade=True, label="Extremely Positive", color = "lightblue")
sns.kdeplot(df[df['Sentiment'] == "Extremely Negative"]['len'], shade=True, label="Extremely Negative", color = "tomato")
sns.kdeplot(df[df['Sentiment'] == "Positive"]['len'], shade=True, label="Positive", color = "lightgreen")
sns.kdeplot(df[df['Sentiment'] == "Negative"]['len'], shade=True, label="Negative", color = "orange")
sns.kdeplot(df[df['Sentiment'] == "Neutral"]['len'], shade=True, label="Neutral", color = "gray")
plt.legend(fontsize=8)
plt.title("Density of Tweet Length by Sentiment")
plt.xlabel("Tweet Length")
plt.ylabel("Density")
plt.savefig("/Users/alexxu/Desktop/UT_Course/SDS384_Scientific_ML/density_tweet_length.pdf", bbox_inches='tight')
```

```python
cnt = pd.DataFrame(df['Sentiment'].value_counts())
```

```python
plt.figure(figsize=(5, 5))
colors = ["lightgreen", "orange", "gray", "lightblue", "tomato"]
sns.barplot(x=cnt['Sentiment'], y=cnt.index, palette=colors)
plt.title("Number of Tweets by Sentiment")
plt.xlabel("Number of Tweets")
plt.ylabel("Sentiment")
plt.savefig("/Users/alexxu/Desktop/UT_Course/SDS384_Scientific_ML/num_tweets_sentiment.pdf", bbox_inches='tight')
```
